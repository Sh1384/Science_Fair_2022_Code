```python
import numpy as np
import pandas as pd
```

```python
icd_codes = pd.read_csv("/cellar/users/mpagadal/data/ukbb/phenotypes/icd.df.melt.tsv", delimiter="\t")
icd_codes = icd_codes.rename(columns={"f.eid":"FID_Salem"})
icd_codes
```

```python
map_files = pd.read_csv("/cellar/users/mpagadal/data/ukbb/mapping/UKBiobank_Salem_shared_ID_mapping_file.txt", delim_whitespace=True)
map_files
```

```python
pheno = pd.merge(icd_codes, map_files, on="FID_Salem")
del pheno["instance"]
del pheno["age_instance"]
del pheno["IID"]
pheno
```

```python
genotypes = pd.read_csv("/cellar/users/mpagadal/projects/tad-meqtl/data/dosage/ukbb.tad.meqtl.raw", delim_whitespace=True)
genotypes
```

```python
genotypes = genotypes.dropna()
genotypes = genotypes.reset_index(drop=True)
genotypes
```

```python
combined = pd.merge(pheno, genotypes, on="FID")
combined
```

```python
print(len(pd.unique(combined["FID_Salem"])))
```

```python
labels = []
for i, row in combined.iterrows():
    if row["cancer ICD10"][0] == "C":
        labels.append(1)
    else:
        labels.append(0)
combined["Labels"] = labels
combined
```

```python
combined.to_csv("risk_cancer_meqtl.csv")
```

```python
snps = []
for col in combined:
    if ":" in col:
        l = col.split(":")
        last = l[-1].split("_")
        snps.append(l[0]+":"+l[1]+":"+l[2]+":"+last[0])
        snps.append(l[0]+":"+l[1]+":"+last[0]+":"+l[2])
snps
```

```python
risk_snps = pd.DataFrame()
risk_snps["SNP"] = snps
risk_snps
```

```python
meqtls = pd.read_csv("meqtls_tads.csv")
meqtls_short = pd.DataFrame()
meqtls_short["SNP"] = meqtls["SNP"]
meqtls_short["TAD_Type"] = meqtls["TAD Type"]
meqtls
```

```python
risk_snps = pd.merge(risk_snps, meqtls_short, on="SNP")
risk_snps
```

```python
risk_snps["TAD_Type"].value_counts()
```

```python
active_genotype = pd.DataFrame()
inactive_genotype = pd.DataFrame()
boundary_genotype = pd.DataFrame()
for col in genotypes:
    if (":" in col):
        l = col.split(":")
        last = l[-1].split("_")
        snp1 = l[0]+":"+l[1]+":"+l[2]+":"+last[0]
        snp2 = l[0]+":"+l[1]+":"+last[0]+":"+l[2]
        if snp1 in list(risk_snps[risk_snps["TAD_Type"] == "Active"]["SNP"]) or snp2 in list(risk_snps[risk_snps["TAD_Type"] == "Active"]["SNP"]):
            active_genotype[col] = genotypes[col]
        elif snp1 in list(risk_snps[risk_snps["TAD_Type"] == "Inactive"]["SNP"]) or snp2 in list(risk_snps[risk_snps["TAD_Type"] == "Inactive"]["SNP"]):
            inactive_genotype[col] = genotypes[col]
        elif snp1 in list(risk_snps[risk_snps["TAD_Type"] == "Boundary"]["SNP"]) or snp2 in list(risk_snps[risk_snps["TAD_Type"] == "Boundary"]["SNP"]):
            boundary_genotype[col] = genotypes[col]
    else:
        active_genotype[col] = genotypes[col]
        inactive_genotype[col] = genotypes[col]
        boundary_genotype[col] = genotypes[col]
```

```python
active_genotype
```

```python
active_genotype.to_csv("modeling_risk/risk_active.csv")
inactive_genotype.to_csv("modeling_risk/risk_inactive.csv")
boundary_genotype.to_csv("modeling_risk/risk_boundary.csv")
```

```python
combined_active = pd.merge(pheno, active_genotype, on="FID")
labels = []
for i, row in combined_active.iterrows():
    if row["cancer ICD10"][0] == "C":
        labels.append(1)
    else:
        labels.append(0)
combined_active["Labels"] = labels
combined_active
```

```python
combined_inactive = pd.merge(pheno, inactive_genotype, on="FID")
labels = []
for i, row in combined_inactive.iterrows():
    if row["cancer ICD10"][0] == "C":
        labels.append(1)
    else:
        labels.append(0)
combined_inactive["Labels"] = labels
combined_inactive
```

```python
combined_boundary = pd.merge(pheno, boundary_genotype, on="FID")
labels = []
for i, row in combined_boundary.iterrows():
    if row["cancer ICD10"][0] == "C":
        labels.append(1)
    else:
        labels.append(0)
combined_boundary["Labels"] = labels
combined_boundary
```

```python
combined_active.to_csv("modeling_risk/icd10_risk_active_meqtls.csv")
combined_inactive.to_csv("modeling_risk/icd10_risk_inactive_meqtls.csv")
combined_boundary.to_csv("modeling_risk/icd10_risk_boundary_meqtls.csv")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import statsmodels.stats.multitest as multi
import matplotlib.pyplot as plt
import os
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
def fdr(df, p):
    probabilities = df[p].to_numpy()
    report = multi.multipletests(probabilities,
                alpha=0.05,
                method="fdr_bh",
                is_sorted=False,
            )
    discoveries = report[1]
    df["fdr"]=discoveries
    df["fdr"]=pd.to_numeric(df["fdr"])
    df["-log10(fdr)"]=-np.log10(df["fdr"])
    return(df)
```

```python
compiled_ukbb = pd.read_csv("risk_score/compiled_ukbb.tsv", delimiter="\t")
compiled_ukbb
```

```python
compiled_ukbb = fdr(compiled_ukbb, "Var1_Pval")
compiled_ukbb = compiled_ukbb[compiled_ukbb["fdr"]<0.1]
compiled_ukbb
```

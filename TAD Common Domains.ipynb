```python
TAD_lengths = [len(GM12878_TAD), len(HMEC_TAD), len(HUVEC_TAD), len(IMR90_TAD), len(NHEK_TAD)]
TAD_lengths
```

```python
TAD_ordered = [GM12878_TAD, IMR90_TAD, NHEK_TAD, HUVEC_TAD, HMEC_TAD]

curr_TADs = pd.DataFrame()
for i in range(len(TAD_ordered)):
    if i < len(TAD_ordered)-1:
        if i == 0:
            TAD_1 = TAD_ordered[i]
            TAD_2 = TAD_ordered[i+1]
        else:
            TAD_1 = curr_TADs
            TAD_2 = TAD_ordered[i+1]
    temp_TAD = pd.DataFrame()
    for j, row in TAD_2.iterrows():
        common_chr = TAD_1[TAD_1["chr1"] == row["chr1"]]
        common_chr.loc[:, "distances"] = np.sqrt((common_chr["x1"] - row["x1"])**2 + (common_chr["x2"] - row["x2"])**2)
        common_TADs = common_chr[common_chr["distances"] <= (min(0.2*abs(row["x1"] - row["x2"]), 50000))]
        temp_TAD = temp_TAD.append(common_TADs, ignore_index = True)

    curr_TADs = temp_TAD
    
    
curr_TADs
```

```python
curr_TADs.to_csv("Carter Lab Datasets/common_TADs.csv")
```

```python
TAD_names = ["GM12878_TAD", "IMR90_TAD", "NHEK_TAD", "HUVEC_TAD", "HMEC_TAD"]
for i in range(len(TAD_ordered)):
    if i != 0:
        TAD1 = TAD_ordered[0]
        TAD2 = TAD_ordered[i]
        temp_TAD1 = pd.DataFrame()
        for j, row in TAD2.iterrows():
            common_chr1 = TAD1[TAD1["chr1"] == row["chr1"]]
            common_chr1.loc[:, "distances"] = np.sqrt((common_chr1["x1"] - row["x1"])**2 + (common_chr1["x2"] - row["x2"])**2)
            common_TADs1 = common_chr1[common_chr1["distances"] <= (min(0.2*abs(row["x1"] - row["x2"]), 50000))]
            temp_TAD1 = temp_TAD1.append(common_TADs1, ignore_index = True)
        venn2(subsets = (len(TAD1)-len(temp_TAD1), len(TAD2)-len(temp_TAD1), len(temp_TAD1)), set_labels = (TAD_names[0], TAD_names[i]))
        plt.show()
    
```
